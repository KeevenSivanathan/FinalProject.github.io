### Joining the Datasets

```python
dc_clean = dc_clean.rename(columns={'YEAR':'Year'})
```

```python
join = [dc_clean,marvel_clean]
```

```python
appearances = pd.concat(join)
```

```python
appearances['Year'] = appearances['Year'].astype(int)
appearances['APPEARANCES'] = appearances['APPEARANCES'].astype(int)
```

```python
appearances = appearances.sort_values(by='APPEARANCES', ascending=False)
```

```python
appearances['year'] = appearances['Year'] // 10 * 10
```

```python
appearances = appearances.drop(['Year'],axis=1)
```

```python
appearances = appearances[appearances['APPEARANCES'] >= 500]
```

```python
appearances
```

```python
#Getting the top 10 values
```

```python
top10 = appearances.nlargest(10,columns='APPEARANCES')
top10
```

```python
top10.to_csv(data_dir + 'top10.csv', index = False)
```
