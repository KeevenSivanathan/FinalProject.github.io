```python
count = powers.apply(pd.value_counts)
```

```python
count = count.iloc[-2:]
```

```python
#Dropping unnecessary columns from data frame
count = count.drop(['hero_names'], axis = 1)
count = count.drop([False], axis = 0)
```

```python
count
```

```python
count = count.T
```

```python
count = count.rename(columns={True: "count"})
```

```python
count = count.sort_values(by=['count'],ascending=False)
```

```python
count['weight'] = 1
```

```python
count
```

```python
count.to_csv(data_dir + 'powers.csv',index=True)
```
