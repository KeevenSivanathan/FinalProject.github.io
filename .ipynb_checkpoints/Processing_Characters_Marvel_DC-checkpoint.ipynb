```python
dc_clean
```

### Joining the DC and Marvel Datasets

```python
join = [dc_clean,marvel_clean]
```

```python
dc_marvel = pd.concat(join)
```

```python
dc_marvel['ALIGN'] = dc_marvel['ALIGN'].str.replace('Good Characters','Good')
dc_marvel['ALIGN'] = dc_marvel['ALIGN'].str.replace('Bad Characters','Bad')
dc_marvel['ALIGN'] = dc_marvel['ALIGN'].str.replace('Neutral Characters','Neutral')
```

```python
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Male Characters','Male')
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Female Characters','Female')
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Genderless Characters','Genderless')
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Transgender Characters','Transgender')
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Genderfluid Characters','Genderfluid')
dc_marvel['SEX'] = dc_marvel['SEX'].str.replace('Agender Characters','A-Gender')
```

```python
dc_marvel['ALIVE'] = dc_marvel['ALIVE'].str.replace('Living Characters','Living')
dc_marvel['ALIVE'] = dc_marvel['ALIVE'].str.replace('Deceased Characters','Deceased')
```

```python
dc_marvel = dc_marvel.drop(['page_id'], axis = 1)
```

```python
#Removing Characters who have a neutral alignment
dc_marvel = dc_marvel[dc_marvel['ALIGN'] != 'Neutral']
```
