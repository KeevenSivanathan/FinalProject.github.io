### Cleaning 

```python
stats.isna().sum() 
```

```python
#Dropping unnecessary columns from data frame
stats_clean = stats.drop(['Unnamed: 0','Unnamed: 0.1','Url','Creator','Hair color','Eye color',
                          'Occupation','Alter Egos','Aliases','Place of birth','Skin color',
                         'First appearance','Team Affiliation','Relatives','Race','Base','Full name',
                          'Total Power','Height','Weight','Gender','Alignment'], axis = 1)
```

```python
stats_clean
```

```python
#For characters which do not have a gender --> Unknown

#stats_clean['Gender'] = stats_clean['Gender'].str.replace('-','Unknown')
```

```python
#For characters with no alignment --> get put to Neutral

#stats_clean['Alignment'] = stats_clean['Alignment'].str.replace('-','neutral')
```

```python
#Filling missing values for the statistics Intelligence, Strength, Speed, Durability, Power, Combat 
stats_clean['Intelligence'].fillna(stats_clean['Intelligence'].median(), inplace=True)
stats_clean['Strength'].fillna(stats_clean['Strength'].median(), inplace=True)
stats_clean['Speed'].fillna(stats_clean['Speed'].median(), inplace=True)
stats_clean['Durability'].fillna(stats_clean['Durability'].median(), inplace=True)
stats_clean['Power'].fillna(stats_clean['Power'].median(), inplace=True)
stats_clean['Combat'].fillna(stats_clean['Combat'].median(), inplace=True)
```

```python
#Getting the average of the 6 different features to create a new feature called 'Overall'
stats_clean['Overall'] = stats_clean.mean(numeric_only=True, axis=1)

#Rounding overall to 1 decimal place
stats_clean['Overall'] = stats_clean['Overall'].round(1)
```

```python
#Statistic of Character dataset fully cleaned
stats_clean 
```

```python
stats_clean = stats_clean.sort_values(by=['Overall'],ascending=False)
```

```python
#This dataset to be used for the interactive bar chart
stats_clean.to_csv(data_dir + 'statistics_full.csv', index = False)
```
